## Cleaning Remaining Null Values
The `rate` column had a significant number of missing values (~15%). Since ratings are subjective and a critical feature for analysis, filling them with the mean or median would introduce bias and distort the true distribution. This leaves us with a cleaner, more reliable dataset for visualization and exploration.

```python
df.dropna(subset=['location','cuisines'],inplace=True)
```

```python
df['rest_type']=df['rest_type'].fillna('Other')
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(',','').astype(float)
df['approx_cost(for two people)']=df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].median())
```

```python
df.dropna(subset=['rate'],inplace=True)
```

Converting to numeric

```python
df=df[~df['rate'].isin(['NEW','-'])].copy()
df['rate']=df['rate'].astype(str).str.replace('/5','',regex=False).str.strip()
df['rate']=df['rate'].astype(float)
```

```python
df.isna().sum()
```

# EDA(Exploratory Data Analysis)
## Distribution of Restaurant types

Understanding the most common types of restaurants listed helps in identifying food industry trends in Bangalore. For instance, a high number of “Quick Bites” or “Cafes” might reflect fast-paced urban lifestyles and changing eating habits.

We'll plot the **top 10 most frequent restaurant types** in the dataset.

```python
top_rest_type=df['rest_type'].value_counts().head(10)
top_rest_type
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=top_rest_type.values,y=top_rest_type.index,hue=top_rest_type.index,palette='viridis')
plt.title('Top 10 restaurants',fontsize=12)
plt.xlabel('No of restaurants')
plt.ylabel('Restaurant type')
plt.tight_layout()
plt.show()
```

- **Quick Bites** and **Casual Dining** dominate Bangalore's food scene, indicating a strong preference for informal, fast-service eateries.
- Presence of **Cafes** and **Dessert Parlours** suggests a thriving youth-oriented and dessert culture.
- Lower representation of niche categories like **Bakeries** or **Fine Dining** shows they’re less mainstream or regionally focused.

This information can help businesses understand **which restaurant formats** are most successful, especially for startups or food delivery platforms.


## Online ordering availability

Understanding how many restaurants offer online ordering helps assess the digital readiness of food businesses in Bangalore. It’s especially useful for food delivery startups and investors to know how saturated the market is.

We'll create a **pie chart** to show the proportion of restaurants that offer online orders.


```python
online_order_counts=df['online_order'].value_counts()
online_order_counts
```

```python
plt.figure(figsize=(8,4))
plt.pie(online_order_counts,labels=online_order_counts.index,autopct='%1.1f%%',startangle=90,colors=['steelblue','cyan'])
plt.title('Online Ordering Availability',fontsize=12)
plt.axis('equal')# Makes pie chart a perfect circle
plt.show()
```

- A large proportion of restaurants **do offer online ordering**, which aligns with Bangalore's tech-savvy and urban customer base.
- However, the presence of restaurants not offering online orders may indicate either traditional setups or dine-in focused business models.

This helps food delivery platforms decide where to focus their onboarding efforts.

## Most Popular Cuisines in Bangalore
We'll analyze the top 10 most frequently listed cuisines.


```python
df['cuisines'].value_counts().head(10)
```

```python
from collections import Counter
cuisine_series=df['cuisines'].apply(lambda x:[i.strip() for i in x.split(',')])
all_cuisines=[cuisine for sublist in cuisine_series for cuisine in sublist]
top_cuisines=Counter(all_cuisines).most_common(10)
cuisine_name,cuisine_counts=zip(*top_cuisines)
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x=list(cuisine_counts), y=list(cuisine_name),hue=list(cuisine_name), palette='cubehelix')
plt.title('Top 10 Cuisines in Bangalore', fontsize=13)
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.tight_layout()
plt.show()
```

### Insights:
- **North Indian** cuisine is the most dominant in Bangalore.
- **Chinese** and **South Indian** cuisines also have a strong presence, showing a blend of traditional and fusion preferences.

## Rating vs Restaurant Type

Boxplots help compare how ratings vary across different restaurant types. 

Let’s visualize the spread of ratings by restaurant type.


```python
top_types=df['listed_in(type)'].value_counts().index[:5]# index only gives labels whereas, head gives labels plus count
top_types
```

```python
filtered_type=df[df['listed_in(type)'].isin(top_types)]
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=filtered_type,x='listed_in(type)',y='rate',hue='listed_in(type)',palette='Set2')
plt.title('Rating Distribution by Restaurant Type')
plt.xlabel('Restaurant Type')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

### Rating Distribution by Restaurant Type

We analyzed how customer ratings vary across the top 5 restaurant types:

- **Cafes** and **Drinks & Nightlife** have the **highest median ratings**, indicating consistently positive customer experiences.
- **Delivery** and **Dine-out** restaurants show a **wider spread in ratings**, suggesting variability in food quality or service.
- **Dessert** places also maintain generally high ratings, though with a few outliers.

This analysis helps us understand which formats are most appreciated by customers in Bangalore’s food scene.


# Cost vs Rating

This scatter plot helps us understand whether there's a relationship between the price range of restaurants and their ratings.  
We're plotting cost (x-axis) against rating (y-axis) to look for trends or correlations.


```python
plt.figure(figsize=(8, 5))
plt.scatter(data=df,x='approx_cost(for two people)',y='rate',alpha=0.5,color='teal')
plt.title('Cost vs Rating')
plt.xlabel('Approximate cost for two people')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

### Cost vs Rating

- The majority of restaurants fall in the **₹200–₹1000 range**, and within this range, ratings are **distributed widely** from 2.5 to 4.5.
- **Higher cost does not always correlate with higher ratings** – expensive restaurants do not consistently have better ratings.
- There are many **budget-friendly restaurants with excellent ratings**, which might indicate good value-for-money options.
- Very high-cost restaurants (₹3000+) are fewer and tend to have ratings between **3.5 to 4.5**, with less density.

 **Conclusion:** A higher price does not guarantee better customer satisfaction.


```python
plt.figure(figsize=(6,4))
sns.boxplot(x='online_order', y='rate',hue='online_order', data=df, palette='coolwarm')
plt.title('Online order availability vs Rating')
plt.xlabel('Online order available')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

### Insights
- Restaurants **offering online orders** tend to have a **slightly higher median rating (~3.9)** compared to those that do not (~3.7).
- The **interquartile range (IQR)** for online-order restaurants is more **compressed**, indicating more consistent customer satisfaction.
- **Outliers** are more frequent and more extreme on the lower end for restaurants **without** online ordering.
- This suggests that restaurants with online ordering **generally perform better** in terms of customer ratings.
- **Business Insight**: Implementing online ordering could be a **competitive advantage** and may improve overall customer experience and ratings.


```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='book_table', y='rate',hue='book_table', palette='Set3')
plt.title('Table Booking Availability vs Rating')
plt.xlabel('Table Booking Available')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

### Table Booking vs Rating

- Restaurants that offer **table booking** tend to have a **higher median rating (~4.0)** compared to those that do not (~3.7).
- **Rating consistency** is better among restaurants with table booking (less spread).
- **Restaurants without table booking** show more low-rated outliers, suggesting possible issues with crowding, delays, or poor walk-in experience.
- This indicates that offering **table booking can positively impact customer satisfaction** and review scores.
  
 **Conclusion**: Table booking is likely a convenience feature that enhances user experience and encourages higher ratings.
